## Задание 1

Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

- если источник traffic_source равен yandex или google, то в source_type ставится organic
- для источников paid и email из России - ставим ad
- для источников paid и email не из России - ставим other
- все остальные варианты берем из traffic_source без изменений

```python
import pandas as pd
```

```python
log = pd.read_csv('visit_log.csv', sep = ';')
log.head()
```

```python
# напишем функцию, которая будет проверять на соответствие условиям
def source_type(row):
    if (row['traffic_source'] == 'yandex') | (row['traffic_source'] == 'google'):
        return 'organic'
    elif (row['traffic_source'] == 'Russia') & ((row['traffic_source'] == 'paid') | (row['traffic_source'] == 'email')):
        return 'ad'
    elif (row['traffic_source'] != 'Russia') & ((row['traffic_source'] == 'paid') | (row['traffic_source'] == 'email')):
        return 'other'
    return row['traffic_source']
```

```python
# применим данную функцию
log['source_type'] = log.apply(source_type, axis = 1)
```

```python
log.head(5)
```

Решим задачу с помощью метода loc:

```python
log.loc[(log['traffic_source'] == 'yandex') | (log['traffic_source'] == 'google'), 'source_type'] = 'organic'
log.loc[((log['traffic_source'] == 'paid') | (log['traffic_source'] == 'email')) & (log['region'] == 'Russia'), 'source_type'] = 'ad'
log.loc[((log['traffic_source'] == 'paid') | (log['traffic_source'] == 'email')) & (log['region'] != 'Russia'), 'source_type'] = 'other'
log.loc[log.source_type.isnull(), 'source_type'] = log['traffic_source']
```

## Задание 3

Используйте файл с оценками фильмов ml-latest-small/ratings.csv. Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.

```python
ratings = pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
# Посчитаем количество оценок
ratings_count = ratings.groupby('userId').count().reset_index()[['userId', 'rating']]
```

```python
ratings_count.head(5)
```

```python
# Выведем минимальное и максимальное значение столбца timestamp
rating_min_max = ratings.groupby('userId').agg(['min', 'max'])['timestamp']
rating_min_max.head(5)
```

```python
# Присоединим к таблице с минимальными и максимальными значениями столбца timestamp количество оценок
rating_result = rating_min_max.join(ratings_count, how = 'left')
rating_result.head()
```

```python
# Выведем пользователей, у которых более 100 оценок
rating_more_than_100 = rating_result[rating_result['rating'] > 100]
rating_more_than_100.head(5)
```

```python
# Поститаем среднее время жизни пользователей:
rating_more_than_100['avg_lifetime'] = rating_more_than_100['max'] - rating_more_than_100['min']
```

```python
rating_more_than_100.head(5)
```

## Задание 4

Дана статистика услуг перевозок клиентов компании по типам:
- rzd - железнодорожные перевозки
- auto - автомобильные перевозки
- air - воздушные перевозки
- client_base - адреса клиентов

Необходимо сформировать две таблицы:
- таблицу с тремя типами выручки для каждого client_id без указания адреса клиента
- аналогичную таблицу по типам выручки с указанием адреса клиента

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
auto = pd.DataFrame(
    {
        'client_id': [113, 114, 115, 116, 117],
        'auto_revenue': [57483, 83, 912, 4834, 98]
    }
)
auto
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```

```python
client_base = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115, 116, 117, 118],
        'address': ['Комсомольская 4', 'Энтузиастов 8а', 'Левобережная 1а', 'Мира 14', 'ЗЖБИиДК 1', 
                    'Строителей 18', 'Панфиловская 33', 'Мастеркова 4']
    }
)
client_base
```

```python
# Объединим все таблицы в одну:
table = rzd.merge(auto, how = 'outer', on = 'client_id')
table = table.merge(air, how = 'outer', on = 'client_id')
table
```

```python
# В таблице заменим значение NaN в столбцах с выручкой на 0
table.loc[table.rzd_revenue.isnull(), 'rzd_revenue',] = 0
table.loc[table.auto_revenue.isnull(), 'auto_revenue',] = 0
table.loc[table.air_revenue.isnull(), 'air_revenue',] = 0
```

```python
table
```

```python
# Присоединим столбцец с адресом:
full_table = table.merge(client_base, how='outer', on='client_id')
full_table.head(5)
```
